```python

import pandas as pd
from os import listdir
from os.path import isfile, join
```

```python
onlyfiles = [f for f in listdir("gameScores") if isfile(join("gameScores", f))]
onlyfiles
```

```python
onlyfiles2 = [f for f in listdir("gameStatistics") if isfile(join("gameStatistics", f))]
onlyfiles2[0]
```

```python

for i in range(len(onlyfiles)):
    name = onlyfiles[i].split("-")
    dataScore = pd.read_csv("gameScores/"+onlyfiles[i])
    dataStats = pd.read_csv("gameStatistics/"+onlyfiles2[i])
    dataStats = dataStats.drop(["market","name","alias","Unnamed: 0","id"], axis=1)
    dataStats = dataStats.iloc[1:]
    dataStats = dataStats.reset_index() 
    result = pd.concat([dataScore,dataStats], axis=1)
   
    result.to_csv("gamesFiltered/"+name[0]+"-season"+name[1]+".csv")
```

```python

onlyfiles3 = [f for f in listdir("gamesFiltered") if isfile(join("gamesFiltered", f))]
```

```python
onlyfiles3 = [file for file in onlyfiles3 if "2008" in file ]
onlyfiles3
```

```python
newDf = pd.DataFrame();
for file in onlyfiles3:
    df = pd.read_csv("gamesFiltered/"+file)
    col = df.columns
    temp = pd.DataFrame();
    away_points=0
    home_points=0
    for i, row in df.iterrows():
        print(file)
        for c in col:
            if c not in "team":
                temp[c+"_"+row["team"]+"_team"] = [row[c]]
        if "home" in row["team"]:
            home_points=row["points"]
        else:
            away_points=row["points"]
        if away_points > home_points:
            temp["winner"] = ["away"]
        elif away_points < home_points:
            temp["winner"] = ["home"]
        else:
            temp["winner"] = ["tie"]
    newDf = newDf.append(temp)
    
    
    
    
newDf.to_csv("final/gamesFinal2008.csv")
```
